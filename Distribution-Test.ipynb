```python
import pandas as pd
import numpy as np
import scipy.stats as scp
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
#Question 1
#simulation random generation

#total number of victims
q1_totalvictims = np.random.triangular(20,80,300,5000)

#number of victims for each hospital
q1_tv_BIM = 0.3*q1_totalvictims 
q1_tv_TM = 0.15*q1_totalvictims 
q1_tv_MG = 0.2*q1_totalvictims 
q1_tv_BM = 0.25*q1_totalvictims 
q1_tv_BW = 0.1*q1_totalvictims 

#transport time (hours) per victim for each hospital
q1_time_BIM = np.random.exponential((7/60),5000) 
q1_time_TM = np.random.exponential((10/60),5000)
q1_time_MG = np.random.exponential((15/60),5000)
q1_time_BM = np.random.exponential((15/60),5000)
q1_time_BW = np.random.exponential((20/60),5000)
```

```python
#question 1 (i)
#average number of victims for each hospital
a1 = np.mean(q1_tv_BIM)
a2 = np.mean(q1_tv_TM)
a3 = np.mean(q1_tv_MG)
a4 = np.mean(q1_tv_BM)
a5 = np.mean(q1_tv_BW)

print ('BIM:%.2f' % a1 + '  TM:%.2f' % a2 + '  MG:%.2f' % a3 + '  BM:%.2f' % a4 + '  BW:%.2f' % a5)
```

```python
#question 1 (ii)

#total transport time for each hospital
q1_totaltime_BIM = [x*y for x,y in zip(q1_time_BIM, q1_tv_BIM)]
q1_totaltime_TM = [x*y for x,y in zip(q1_time_TM, q1_tv_TM)]
q1_totaltime_MG = [x*y for x,y in zip(q1_time_MG, q1_tv_MG)]
q1_totaltime_BM = [x*y for x,y in zip(q1_time_BM, q1_tv_BM)]
q1_totaltime_BW = [x*y for x,y in zip(q1_time_BW, q1_tv_BW)]

#average total transport time for each hospital
b1 = np.mean(q1_totaltime_BIM)
b2 = np.mean(q1_totaltime_TM)
b3 = np.mean(q1_totaltime_MG)
b4 = np.mean(q1_totaltime_BM)
b5 = np.mean(q1_totaltime_BW)

print ('BIM:%.2f' % b1 + '  TM:%.2f' % b2 + '  MG:%.2f' % b3 + '  BM:%.2f' % b4 + '  BW:%.2f' % b5)
```

```python
print ('Question 1')
print ('i)  average number of victims for each hospital: ')
print ('    BIM:%.2f' % a1 + '  TM:%.2f' % a2 + '  MG:%.2f' % a3 + '  BM:%.2f' % a4 + '  BW:%.2f' % a5)
print ('ii) average total transport time for each hospital')
print ('    BIM:%.2f' % b1 + '  TM:%.2f' % b2 + '  MG:%.2f' % b3 + '  BM:%.2f' % b4 + '  BW:%.2f' % b5)
```

```python
#question 1 (iii)
#law of large numbers plot
q1_total_samplemean = np.mean(q1_totalvictims)

cumv = np.cumsum(q1_totalvictims)
avgv = [x/y for x,y in zip(cumv,list(np.arange(1,5001)))]
df = pd.DataFrame({ 'Avg Victims' : avgv}) #define data frame

df.plot(title="Law of Large Numbers for Total")
plt.ylim({50,250})
plt.xlabel('Sample Size')
plt.ylabel('Mean of number of Total Victims')
plt.axhline(y=np.mean(q1_totalvictims),color='r',linewidth = 0.6)
plt.show()

print ('         Sample Mean(red line) : %.4f' %q1_total_samplemean ) 
```

```python
#law of large numbers plot for each hospital
#BIM
plt.subplot(2,3,1)
cumv_BIM = np.cumsum(q1_tv_BIM)
avgv_BIM = [x/y for x,y in zip(cumv_BIM,list(np.arange(1,5001)))]
df1 = pd.DataFrame({ 'Avg Victims' : avgv_BIM}) #define data frame

plt.plot(df1)
plt.ylim({15,65})
plt.axhline(y=a1,color='r',linewidth = 0.8)
plt.title('BIM Mean:%.2f' %a1)

#TM
plt.subplot(2,3,2)
cumv_TM = np.cumsum(q1_tv_TM)
avgv_TM = [x/y for x,y in zip(cumv_TM,list(np.arange(1,5001)))]
df2 = pd.DataFrame({ 'Avg Victims' : avgv_TM}) #define data frame

plt.plot(df2)
plt.ylim({15,25})
plt.axhline(y=a2,color='r',linewidth = 0.8)
plt.title('TM Mean:%.2f' %a2)

#MG
plt.subplot(2,3,3)
cumv_MG = np.cumsum(q1_tv_MG)
avgv_MG = [x/y for x,y in zip(cumv_MG,list(np.arange(1,5001)))]
df3 = pd.DataFrame({ 'Avg Victims' : avgv_MG}) #define data frame

plt.plot(df3)
plt.ylim({15,35})
plt.axhline(y=a3,color='r',linewidth = 0.8)
plt.title('MG Mean:%.2f' %a3)

#BM
plt.subplot(2,3,4)
cumv_BM = np.cumsum(q1_tv_BM)
avgv_BM = [x/y for x,y in zip(cumv_BM,list(np.arange(1,5001)))]
df4 = pd.DataFrame({ 'Avg Victims' : avgv_BM}) #define data frame

plt.plot(df4)
plt.ylim({15,45})
plt.axhline(y=a4,color='r',linewidth = 0.8)
plt.title('BM Mean:%.2f' %a4)

#BW
plt.subplot(2,3,5)
cumv_BW = np.cumsum(q1_tv_BW)
avgv_BW = [x/y for x,y in zip(cumv_BW,list(np.arange(1,5001)))]
df5 = pd.DataFrame({ 'Avg Victims' : avgv_BW}) #define data frame

plt.plot(df5)
plt.ylim({5,20})
plt.axhline(y=a5,color='r',linewidth = 0.8)
plt.title('BW Mean:%.2f' %a5)

plt.subplots_adjust(wspace=0.8,hspace=0.8)
plt.show()
```
